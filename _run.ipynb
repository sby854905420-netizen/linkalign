```python
import json 
import pandas as pd
import random
```

```python
path = "/home/xubeiyu/projects/LinkAlign/spider2_dev/spider2_dev_preprocessed.json"

data_df = pd.read_json(path)
```

```python
samples_df = data_df.sample(3).copy()
all_db_ids = list(set(data_df['db_id']))
if len(all_db_ids) < 3:
    raise ValueError(f"Need at least 3 unique db_id values, got {len(all_db_ids)}")
```

```python
def build_candidate_db_ids(true_db_id):
    noise_pool = [db for db in all_db_ids if db != true_db_id]
    noise_db_ids = random.sample(noise_pool, 2)
    candidate_db_ids = [true_db_id] + noise_db_ids
    random.shuffle(candidate_db_ids)
    return candidate_db_ids

samples_df['candidate_db_ids'] = samples_df['db_id'].apply(build_candidate_db_ids)
```
